```python
#3.2 EWO Revision 1 Loops
#EWO = Elliott Wave Oscillator
#The current code only focuses on the long position
#Tested Ranges - Fast SMA: 2~10 days, Periods to define a slope of EWO: 3~5 days, Slow SMA: 20~50 days

import import_ipynb
import EWO

#Dataframe Initialization
#target_market_ticker = EWO.Ewo(target_market_ticker, start_date(yyyy-mm-dd), end_date(yyyy-mm-dd))
SPY = EWO.Ewo_loops1('SPY', '2010-06-28', '2019-06-28')

for n in range(2, 11): #Fast SMA: 2~10 days
    for l in range(3, 6): #Periods to define a slope of EWO: 3~5 days
        for m in range(20, 51): #Slow SMA: 20~50 days
            #Import and process data & return market and market_buy_init dataframes
            #market, market_buy_init = target_market_ticker.data_process(fast_sma, slow_sma, \
            #longer_sma, n_day_EWO_slope, n_day_closing_price, use_filter4(0 or 1))
            market, market_buy_init = SPY.data_process(n, m, 50, l, 50, 0)

            #Backtesting of trading strategy on the target_market_ticker from start_date to end_date
            #Return budget and signal_num
            budget, signal_num = SPY.backtest(100000, market, market_buy_init)
            
            #Append budget & signal_num of each combination to the corresponding lists
            #Print combinations tested so far
            SPY.collect(budget, signal_num, n, l)

#Store max_profit & max_trade and print the backtesting result
#Return profit & trade_num dataframes
profit, trade_num = SPY.data_process2()
```
